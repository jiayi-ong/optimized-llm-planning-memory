# 01 — Explore the Travel Simulator

This notebook walks through the SimulatorAdapter API:
- Instantiate a world with a fixed seed
- Search flights, hotels, and activities
- Inspect the Pydantic output schemas
- Verify world reproducibility across seeds

```python
import sys
sys.path.insert(0, '../src')

from optimized_llm_planning_memory.simulator.adapter import SimulatorAdapter

sim = SimulatorAdapter(seed=42)
print('Simulator created. Seed:', sim.get_world_seed())
```

```python
# Search flights
flights = sim.search_flights(
    origin='New York', destination='Paris',
    date='2025-06-01', num_passengers=2
)
print(f'Found {len(flights)} flights')
for f in flights[:3]:
    print(f)
```

```python
# Search hotels
hotels = sim.search_hotels(
    city='Paris', check_in='2025-06-01',
    check_out='2025-06-04', num_guests=2
)
print(f'Found {len(hotels)} hotels')
for h in hotels[:3]:
    print(h)
```

```python
# Verify seed reproducibility
sim2 = SimulatorAdapter(seed=42)
flights2 = sim2.search_flights('New York', 'Paris', '2025-06-01', 2)
assert flights == flights2, 'Same seed should produce identical results'
print('Reproducibility check passed.')
```
